```python
import torch
import pandas as pd
from cleanlab.dataset import health_summary
```

```python
vgg16_df = pd.read_json('tobacco3482-vgg16-cleanlab.json')
```

```python
labels = ["ADVE", "Email", "Form", "Letter", "Memo", "News", "Note", "Report", "Resume", "Scientific"]
id2label = {v: k for v, k in enumerate(labels)}
label2id = {v: k for k, v in id2label.items()}
```

```python
noisy_vgg16_labels = [label2id[i] for i in vgg16_df['true_label'].to_list()]
```

```python
pred_probs_vgg16 = torch.Tensor(vgg16_df['logits'].to_list())
pred_probs_vgg16 = torch.nn.Softmax(1)(pred_probs_vgg16).numpy()
```

```python
health_summary(labels=noisy_vgg16_labels,
               pred_probs=pred_probs_vgg16)
```
